```python
# calcular_valor_inventario(self): Calcula y muestra el valor total del inventario.
#Utiliza un bucle for para calcular el valor total del inventario.
#Itera a través del inventario y suma el valor de cada producto (precio x cantidad).
#  Es decir,calcula el valor total del inventario. 
# Ejemplo:
# si tenemos 5 camisas que valen 5 euros
# y 10 calcetines que valen 1 euro
# este método te tiene que devolver: 35 euros
```

```python
# el planteamiento es el siguiente, de cada diccionario , dentro de inventario ,tienen que multiplicarse 
#la segunda (cantidad)y tercera (precio) key_value (en humano)
#y luego sumarse cada una de los diccionarios multiplicados 
#

def calcular_valor_inventario(self):
    valor_total=0 #variable vacia que recibira la multiplicacion de cantidad y precio

    #iteramos 
    for producto in self.inventario:

        valor_producto=producto['precio']*producto['cantidad']# como va a iterar, aqui paran cada uno de las multiplicaciones e cada producto

        valor_total+=valor_producto #es la suma de todas las multiplicaciones de valor producto

        print(f'el valor total de todo el "inventario es: {valor_total} €')
        return valor_total

```
